This is only for manipulating and working with datasets in Python (pandas), seperate module for plotting and visualization will be added

```python
import pandas as pd
print(pd.__version__)
```

```python
"""
Loads in dataset from url with json format and prints the five first rows of the dataset.
"""
df = pd.read_json("https://data.smcgov.org/resource/mb6a-xn89.json")
df.head(5)
```

```python
"""
Returns dimensions of the dataset.
"""
df.shape
```

```python
"""
Describe gives us a numerical analysis of each column, f.ex returning min and max value of the column
and nummber of elements.
"""
df.describe()
```

```python
"""
dtypes returns the data type of each row in the dataset.
"""
df.dtypes
```

```python
"""
drop allows us to remove a column, usually we will remove a column that is not helpful to what we're trying to achieve.
describe is added at the end to print out a new description of the modified dataset. include = "all" will include non-numeric
as well in the description
"""
df.describe(include="all")
```

```python
"""
Counts elements in each column.
"""
df.count()
```

```python
"""
By typing in the name of a column before count we can get the number of elements from that column.
"""
df.geography.count()
```

```python
"""
Lists the unique elements in an array.
"""
df.geography_type.unique()
```

```python
"""
Counts how many times each element appears.
"""
df.less_than_high_school_graduate.value_counts()
```
